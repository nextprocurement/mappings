```python
# Initialize an empty list to store column names that are full of null values
null_columns = []

# Iterate through the columns of the DataFrame
for column in source.columns:
    # Create a boolean mask for NaN values in the column
    is_null = source[column].isnull()
    
    # Check if all values in the boolean mask are True
    if is_null.all():
        null_columns.append(column)

# Print the column names that are full of null values
for column in null_columns:
    print(f"Column '{column}' is full of null values.")


# Initialize an empty list to store column names that meet the condition
columns_with_value = []

# Iterate through the columns of the DataFrame
for column in source.columns:
    # Count the occurrences of the value in the column
    count = (source[column].astype(str) == "['nan']").sum()
    
    # Check if the count is the length of source
    if count == len(source):
        columns_with_value.append(column)

# Print the column names that meet the condition
for column in columns_with_value:
    print(f"Column '{column}' is full of nan values.")
```
